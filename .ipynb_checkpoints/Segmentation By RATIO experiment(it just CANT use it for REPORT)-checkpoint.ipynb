```python
def get_ul_coordinat(coordinat, height):
    height_tanwin = coordinat[3] - coordinat[1]
    # upper
    if coordinat[1]-height_tanwin > 0:
        upper = [coordinat[0], coordinat[1]-height_tanwin,
                 coordinat[2], coordinat[3]-height_tanwin]
    else:
        upper = [coordinat[0], 0,
                 coordinat[2], coordinat[3]-height_tanwin]
    # lower
    if coordinat[1]+height_tanwin < height:
        lower = [coordinat[0], coordinat[1]+height_tanwin,
                 coordinat[2], coordinat[3]+height_tanwin]
    else:
        lower = [coordinat[0], coordinat[1]+height_tanwin,
                 coordinat[2], height]
    
    return upper, lower

def upper_or_lower(bw_img, upper, lower):
    upper_count = 0
    for x in range(upper[0], upper[2]):
        for y in range(upper[1], upper[3]):
            if bw_img[y, x] < 1:
                upper_count += 1

    lower_count = 0
    for x in range(lower[0], lower[2]):
        for y in range(lower[1], lower[3]):
            if bw_img[y, x] < 1:
                lower_count += 1

    return upper_count, lower_count

def get_lr_coordinat(coordinat, width):
    width_tanwin = coordinat[2] - coordinat[0]
    if coordinat[0] - width_tanwin > 0:
        left = [coordinat[0]-width_tanwin, coordinat[1],
                coordinat[2]-width_tanwin, coordinat[3]]
    else:
        left = [0, coordinat[1], 
                coordinat[2]-width_tanwin, coordinat[3]]
    if coordinat[2] + width_tanwin < width:
        right = [coordinat[0]+width_tanwin, coordinat[1],
                 coordinat[2]+width_tanwin, coordinat[3]]
    else:
        right = [coordinat[0]+width_tanwin, coordinat[1],
                 width, coordinat[3]]
        
    return left, right
```

```python
bw_img = image.copy()
img_copy = image.copy()
height, width = bw_img.shape
kernel = np.ones((5,5),np.uint8)
img_copy = cv2.morphologyEx(img_copy, cv2.MORPH_OPEN, kernel)
# cv2.imshow('d', img_copy)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
upper, lower = get_ul_coordinat(coordinat, height)
upper_count, lower_count = upper_or_lower(img_copy, upper, lower)

if upper_count < lower_count:
    print('lower')
    cv2.rectangle(bw_img, (lower[0], lower[1]),
                          (lower[2], lower[3]), (100, 150,0), 2)
    first_pos = lower
    first_pos_direction = 'down'
elif upper_count > lower_count:
    print('upper')
    cv2.rectangle(bw_img, (upper[0], upper[1]),
                          (upper[2], upper[3]), (100, 150,0), 2)
    first_pos = upper
    first_pos_direction = 'up'
else:  # if same then expand
    print('expand upper lower checking')
    while(upper_count==lower_count):
        upper, _ = get_ul_coordinat(upper, height)
        _, lower = get_ul_coordinat(lower, height)
        upper_count, lower_count = upper_or_lower(img_copy, upper, lower)
        if upper_count < lower_count:
            print('lower')
            cv2.rectangle(bw_img, (lower[0], lower[1]),
                                  (lower[2], lower[3]), (100, 150,0), 2)
            first_pos = lower
            first_pos_direction = 'down'
        elif upper_count > lower_count:
            print('upper')
            cv2.rectangle(bw_img, (upper[0], upper[1]),
                                  (upper[2], upper[3]), (100, 150,0), 2)
            first_pos = upper
            first_pos_direction = 'up'

# check left and right if no white then split 2 by 2 
# if white then split 3 1 (3 on direction of the opposite tanwin)
left, right = get_lr_coordinat(first_pos, width)
left_count, right_count = upper_or_lower(img_copy, left, right)

if left_count > 1 and right_count > 1:
    print('fp_mid')
    fp_h = first_pos[3] - first_pos[1]
    a = 2
    b = 3
    if first_pos[1] - a * fp_h > 0 and first_pos[1] + b * fp_h < height: 
        oneline_img_coordinat = [0, first_pos[1] - a * fp_h,
                                 width, first_pos[1] + b * fp_h]
    elif first_pos[1] - a * fp_h < 0 and first_pos[1] + b * fp_h < height: 
        oneline_img_coordinat = [0, 0, width, first_pos[1] + b * fp_h]
    elif first_pos[1] - a * fp_h > 0 and first_pos[1] + b * fp_h > height: 
        oneline_img_coordinat = [0, first_pos[1] - a * fp_h,
                                 width, height]
    else:
        oneline_img_coordinat = [0, 0, width, height]

    cv2.rectangle(bw_img, 
                  (oneline_img_coordinat[0], oneline_img_coordinat[1]),
                  (oneline_img_coordinat[2], oneline_img_coordinat[3]),
                  (100, 150,0), 2)

if left_count < 1 or right_count < 1:
    print('fp_off')
    fp_h = first_pos[3] - first_pos[1]
    if first_pos_direction == 'down':
        a = 1
        b = 4
        if first_pos[1] - a * fp_h > 0 and first_pos[1] + b * fp_h < height: 
            oneline_img_coordinat = [0, first_pos[1] - a * fp_h,
                                     width, first_pos[1] + b * fp_h]
        elif first_pos[1] - a * fp_h < 0 and first_pos[1] + b * fp_h < height: 
            oneline_img_coordinat = [0, 0, width, first_pos[1] + b * fp_h]
        elif first_pos[1] - a * fp_h > 0 and first_pos[1] + b * fp_h > height: 
            oneline_img_coordinat = [0, first_pos[1] - a * fp_h,
                                     width, height]
        else:
            oneline_img_coordinat = [0, 0, width, height]

    if first_pos_direction == 'up':
        a = 4
        b = 1
        if first_pos[1] - a * fp_h > 0 and first_pos[1] + b * fp_h < height: 
            oneline_img_coordinat = [0, first_pos[1] - a * fp_h,
                                     width, first_pos[1] + b * fp_h]
        elif first_pos[1] - a * fp_h < 0 and first_pos[1] + b * fp_h < height: 
            oneline_img_coordinat = [0, 0, width, first_pos[1] + b * fp_h]
        elif first_pos[1] - a * fp_h > 0 and first_pos[1] + b * fp_h > height: 
            oneline_img_coordinat = [0, first_pos[1] - a * fp_h,
                                     width, height]
        else:
            oneline_img_coordinat = [0, 0, width, height]


    cv2.rectangle(bw_img, 
                  (oneline_img_coordinat[0], oneline_img_coordinat[1]),
                  (oneline_img_coordinat[2], oneline_img_coordinat[3]),
                  (100, 150,0), 2)
```
